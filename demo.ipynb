```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
```

# Data

```python
iris_arr = datasets.load_iris()
iris = pd.DataFrame(data= np.c_[iris_arr['data'], iris_arr['target']],
                     columns= iris_arr['feature_names'] + ['target'])
```

```python
iris.head()
```

```python
sns.scatterplot(iris, x = 'sepal length (cm)', y= 'sepal width (cm)', hue = 'target')
```

# Train Test Split

```python
X = iris.iloc[:,:-1]
y = iris.iloc[:,-1]
```

```python
print(X.head())
print(y.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
x_train
```

# Modeling

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred_train = knn.predict(x_train)
y_pred_test = knn.predict(x_test)
```

```python
training_accuracy = sum(y_pred_train == y_train) / len(y_train)
testing_accuracy = sum(y_pred_test == y_test) / len(y_test)
print(f"Training Accuracy: {round(training_accuracy*100, 1)}%")
print(f"Testing Accuracy: {round(testing_accuracy*100, 1)}%")
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
y_test_pred = clf.predict(x_test)
y_train_pred = clf.predict(x_train)
```

```python
training_accuracy = sum(y_train_pred == y_train) / len(y_train)
testing_accuracy = sum(y_test_pred == y_test) / len(y_test)
print(f"Training Accuracy: {round(training_accuracy*100, 1)}%")
print(f"Testing Accuracy: {round(testing_accuracy*100, 1)}%")
```
